```python
import pandas as pd
from functools import reduce
data = pd.read_csv("/content/NewspaperData.csv")
total_daily = data.groupby('Newspaper')['daily'].min()
total = reduce(lambda x, y: x * y, total_daily)
print(total)
```

```python
import pandas as pd
from functools import reduce
data = pd.read_csv("/content/Sports data.csv")
total= data.groupby('Year of birth')['Year of birth'].sum()
total= reduce(lambda x, y: x + y, total)
print(total)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/Sports data.csv')
df['Rank']=df['Rank'].apply(lambda x:x*2)
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/Sports data.csv')
plt.hist(data['Salary/Winnings'], bins=20)
plt.title('HISTOGRAM')
plt.xlabel('Salary/Winnings')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/Sports data.csv')
plt.scatter(data['Height (cm)'], data['Rank'])
plt.title('SCATTER PLOT')
plt.xlabel('Height (cm)')
plt.ylabel('Rank')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/Sports data.csv')
data['Gender'].value_counts().plot(kind='bar')
plt.title('BAR CHART')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('/content/Sports data.csv')
plt.figure(figsize=(7,5))
sns.boxplot(x=data['Total Pay'])
plt.title('BOX PLOT')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('/content/Sports data.csv')
plt.figure(figsize=(5,5))
sns.lineplot(x=data['Endorsements'], y=data['Height (cm)'])
plt.title('Height (cm) VS Endorsements')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/Sports data.csv')
plt.figure(figsize=(10,5))
plt.pie(data['Sport'].value_counts(), labels = data['Sport'].unique(),autopct='%1.1f%%')
plt.title('PIE CHART')
plt.show()
```
